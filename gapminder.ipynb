```python
import pandas as pd
from gapminder import gapminder

# Load the data
df = gapminder

# Display the first few rows of the dataset
print(df.head())

# Display basic information about the dataset
print(df.info())

# Summary statistics of the dataset
print(df.describe())
```

GDP per capita is calculated by dividing the GDP by the total population of the country. It is a measure of a country's economic output that accounts for its number of people. It provides an average economic productivity and living standards.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot of GDP per Capita vs. Life Expectancy
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='gdpPercap', y='lifeExp', hue='continent', size='pop', sizes=(20, 200))
plt.xscale('log')
plt.title('GDP per Capita vs. Life Expectancy')
plt.xlabel('GDP per Capita (log scale)')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['lifeExp'], kde=True)
plt.title('Distribution of Life Expectancy')
plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(data=df, x='year', y='pop', hue='continent', estimator='sum')
plt.title('Population Growth Over Time by Continent')
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(data=df, x='year', y='lifeExp', hue='continent', estimator='mean')
plt.title('Life Expectancy Over Time by Continent(Average)')
plt.xlabel('Year')
plt.ylabel('Life Expectancy(Average)')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(data=df, x='year', y='lifeExp', hue='continent', estimator='mean')
plt.title('Life Expectancy Over Time by Continent(Median)')
plt.xlabel('Year')
plt.ylabel('Life Expectancy(Midian)')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='continent', y='gdpPercap')
plt.yscale('log')
plt.title('GDP per Capita Distribution by Continent')
plt.xlabel('Continent')
plt.ylabel('GDP per Capita (log scale)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a histogram for gdpPercap for each continent
plt.figure(figsize=(14, 10))

# List of unique continents
continents = df['continent'].unique()

# Plot histograms for each continent
for i, continent in enumerate(continents):
    plt.subplot(3, 2, i + 1)
    subset = df[df['continent'] == continent]
    sns.histplot(subset['gdpPercap'], kde=True, bins=30)
    plt.xscale('log')
    plt.title(f'{continent}')
    plt.xlabel('GDP per Capita (log scale)')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.suptitle('GDP per Capita Distribution by Continent', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
correlation_matrix = df[['lifeExp', 'pop', 'gdpPercap']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
